```python
data1 = pd.DataFrame(data)
```

# Randomly 70% of the whole data is sampled

```python
data1 = data.sample(n=None, frac=0.7, replace=False, weights=None, random_state=1324, axis=None)
```

```python
data1.head()
```

# Creating Feature and Label

```python
label = data1.iloc[:,30]
```

```python
label.head()
```

# Creating Train Data

```python
feature = data1.drop(['Result'],axis =1)
```

```python
feature.head(2)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test,Y_train,Y_test = train_test_split(feature,label,test_size = 0.2,random_state = 0)
```

```python
X_train.shape, Y_train.shape
```

```python
X_test.shape, Y_test.shape
```

# Creating Model

```python
from sklearn.linear_model import LogisticRegression
clas = LogisticRegression(random_state=0)
result = clas.fit(X_train, Y_train)
result
```

```python
y_predict = clas.predict(X_test)
```

```python
y_predict
```

```python
print (result.intercept_, result.coef_) 
print()
#print(result.score(X_train, Y_train))
```

```python
from sklearn.metrics import confusion_matrix
matrix = confusion_matrix(Y_test,y_predict)
matrix
```

```python
from sklearn import model_selection
from sklearn import metrics, cross_validation
#from sklearn.model_selection import cross_val_score,metrics
kfold = model_selection.KFold(n_splits=10, random_state=7)
modelCV = LogisticRegression()
scoring = 'accuracy'
results = model_selection.cross_val_score(result, X_train, Y_train, cv=kfold, scoring=scoring)
print (metrics.accuracy_score(Y_test, y_predict))
print()
print("10-fold cross validation average accuracy: %.3f" % (results.mean()))
```
