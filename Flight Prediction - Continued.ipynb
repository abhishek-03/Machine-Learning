```python
def convert_nd(x):
    if 'Delhi' in str(x):
        return 'New Delhi'
    else:
        return str(x)
train['Source']=train['Source'].apply(convert_nd)
train['Destination']=train['Destination'].apply(convert_nd)
```

```python
print('Now the unique Source locations are :' , train['Source'].unique())
print()
print('Now the unique Destination locations are :' , train['Destination'].unique())
```

```python
def stop(x):
    if 'non-stop' in str(x):
        x=x.split(' ')
        return 0
    elif 'stop' in str(x):
        x=x.split(' ')
        return int(x[0]*1)
    else:
        return x

train['Total_Stops']=train['Total_Stops'].apply(stop)
```

```python
train=train[~train['Additional_Info'].isin(['Red-eye flight','1 Short layover','2 Long layover'])]
train["Airline"]=train.Airline.replace("Trujet", "SpiceJet")
train["Airline"]=train.Airline.replace('Vistara Premium economy', 'Vistara')
train["Airline"]=train.Airline.replace('Jet Airways Business', 'Jet Airways')
train["Airline"]=train.Airline.replace('Multiple carriers Premium economy', 'Multiple carriers')
train.Additional_Info.replace(['No Info'], ['No info'], inplace=True)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
def routeMap(val):
    val=val.split(" → ")
    val2=[i[:3] for i in val]
    return val2
train['Route']=train['Route'].apply(routeMap)
```

```python
def time_bins(v):
    if str(v)>='00:00' and str(v)<'06:00':
        return 'Morning'
    elif str(v)>='06:00' and str(v)<'12:00':
        return 'Afternoon'
    elif str(v)>='12:00' and str(v)<'18:00':
        return 'Evening'
    elif str(v)>='18:00' and str(v)<='23:60':
        return 'Night'
    else:
        return v
train['Dep_Time']=train['Dep_Time'].apply(time_bins)
```

```python
def daytime(x):
    if 'Morning' in str(x):
        return 1*6
    elif 'Afternoon' in str(x):
        return 2*6
    elif 'Evening' in str(x):
        return 3*6
    else:
        return 4*6

train['Dep_Time']=train['Dep_Time'].apply(daytime)
```

```python
#'Banglore' 'Kolkata' 'New Delhi' 'Chennai' 'Mumbai''Cochin''Hyderabad'

def convert_ds(x):
    if 'New Delhi' in str(x):
        return 7
    elif 'Banglore' in str(x):
        return 5
    elif 'Kolkata' in str(x):
        return 6
    elif 'Chennai' in str(x):
        return 5
    elif 'Mumbai' in str(x):
        return 8
    elif 'Cochin' in str(x):
        return 3
    else:
        return 4

train['Destination']=train['Destination'].apply(convert_ds)
train['Source']=train['Source'].apply(convert_ds)
```
