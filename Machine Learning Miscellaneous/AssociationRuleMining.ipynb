```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
store_data = pd.read_excel("/home/subhasis/Dropbox/Store data.xlsx")
```

```python
store_data.head()
```

```python
store_data['cnt'] = 1
```

```python
basket = (store_data.groupby(['CASH_MEMO','PRODUCT'])['cnt']
          .sum().unstack().reset_index().fillna(0).set_index('CASH_MEMO'))
```

```python
def recode(x):
    if x<=0:
        return 0
    if x>0:
        return 1
basket_bin = basket.applymap(recode)
```

```python
basket_rules = apriori(basket_bin,min_support=0.01, use_colnames=True)
```

```python
association_rules(basket_rules, metric='lift', min_threshold=2).sort_values(by='lift', ascending=False)
```
